```python
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
%matplotlib inline
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras import utils
```

```python
#Делим наш датасэт на обучающую и тестовую выборку
```

```python
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()
```

```python
class_names = ['футболка', 'брюки', 'свитер', 'платье', 'пальто', 'туфли', 'рубашка', 'кроссовки', 'сумка', 'ботинки']
```

```python
#Проводим предварительную обработку данных
```

```python
#Нормализация данных
```

```python
x_train = x_train / 255
x_test = x_test / 255
```

```python
#Просмотр нескольких изображений
```

```python
plt.figure(figsize=(10,10))
for i in range (50):
 plt.subplot(5,10,i+1)
 plt.xticks([])
 plt.yticks([])
 plt.imshow(x_train[i], cmap=plt.cm.binary)
 plt.xlabel(class_names[y_train[i]])
```

```python
#Создание модели нейронной сети
```

```python
model = keras. Sequential([
                keras.layers.Flatten(input_shape=(28,28)),
                keras.layers.Dense(128, activation="relu"),
                keras.layers.Dense(10, activation="softmax")
])
```

```python
#Компиляция модели
```

```python
model.compile(optimizer=tf.keras.optimizers.SGD(), loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
#Обучение модели
```

```python
model.fit(x_train, y_train, epochs=10)
```

```python
#Проверка точности предсказания
```

```python
test_loss, test_acc = model.evaluate(x_test, y_test)
print('Качество предсказаний:', test_acc)
```

```python
#Предсказывание
```

```python
predictions = model.predict(x_train)
```
